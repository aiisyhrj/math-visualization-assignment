```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,300)
y1 = x
y2 = x**2
y3 = np.sin(x)
y4 = np.exp(-0.1* x ) * np.cos(x)

plt.figure(figsize=(8, 5))

plt.plot(x, y1, linestyle="-", label="y = x")
plt.plot(x, y2, linestyle="--", label="y = x²")
plt.plot(x, y3, linestyle=":", label="y = sin(x)")
plt.plot(x, y4, linestyle="-.", marker="o", markevery=15,
         label="y = e^(-0.1x)cos(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("function_plot.png")
plt.show()
```

```python
x = np.linspace(0, 10, 200)

y = np.sin(3*x)+2 # Sine wave with upward shift
plt.plot(x, y, label="y = sin(3x) + 2")
plt.title("y = sin(3x) + 2")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)
plt.savefig("own_equation.png")
plt.show()

```

```python
students = ["S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9", "S10"]
midterm = [85, 72, 90, 66, 78, 92, 60, 74, 88, 95]
final = [80, 70, 94, 68, 75, 90, 65, 72, 84, 96]

total = 0.4 * np.array(midterm) + 0.6 * np.array(final)

#scatter plot
plt.figure(figsize=(8,5))

plt.scatter(midterm, final)

plt.title("Midterm vs Final Scores")
plt.xlabel("Midterm Score")
plt.ylabel("Final Score")
plt.grid(True)

plt.savefig("score_scatter.png")
plt.show()


#histogram
plt.figure(figsize=(8,5))

plt.hist(total, bins=5)

plt.title("Distribution of Total Scores")
plt.xlabel("Total Score")
plt.ylabel("Frequency")
plt.grid(True)

plt.savefig("score_histogram.png")
plt.show()


#bar chart
plt.figure(figsize=(8,5))

plt.bar(students, total)

plt.title("Student Total Scores")
plt.xlabel("Students")
plt.ylabel("Total Score")
plt.grid(True)

plt.savefig("score_bar_chart.png")
plt.show()
```
